Leeo el archivo Json de cars.json creado en la notebook 1_Scrapping_Data.ipynb, y visualizo un pequeno set de datos. Para analizar los datos obtenidos

```python
df = pd.read_json("1_cars.json", orient="columns")
df.tail()
```

Elimino los datos en null porque no me intersan


```python
df2 =df.dropna(axis=1, how='all')
```

```python
df2.tail()
```

Miro la lisa de columnas disponibles, para ver el resultado final sin columnas en null

```python
list(df2.columns) 
```

Comienzo a eliminar los columnas que no son datos relevantes para el modelo , y no incidiran en las predicciones finales.

```python
#id de informte tecnico interno no sirve
df2.pop('codigoInformeTecnico')
#Tipo de combustible tampoco, solo tenemos autos de gasolina
df2.pop('combustible')
#Consumos lo tenemos con valor 0 en todas la filas, lo eliminamos
#df.consumos.unique()
df2.pop('consumos')
#eliminamos detalles , es un texto no especifico
df2.pop('detalle')
#fecha hologacion no nos sirve
df2.pop('fechaHomologacion')
df2.pop('idEtiqueta')
df2.pop('idMarca')
df2.pop('idModelo')
df2.pop('idVehiculo')
df2.pop('nombreEtiqueta')
df2.pop('nombreMarca')
df2.pop('nombreModelo')
#Propulsion todos las filas tiene combustible como valor, no sirve
#df.propulsion.unique()
df2.pop('propulsion')
#Elimino campo vigente interno
#df.vigente.unique()
df2.pop('vigente')
#Elimino year vehicule, todos los datos estan en 0 (cero)
#df.yearVehiculo.unique()
df2.pop('yearVehiculo')
df2.pop('empresaHomologa')

#SE ELIMINARA CATEGORIA PORQUE A LA FECHA NO SE INTERPRETO LO QUE SIGNIFICA.
df2.pop('categoria')


#Se eliminan datos que son importantes pero que en la prediccion son muy dificiles que se consigan para poderpredecir

df2.pop('co2')
df2.pop('rendimientoCarretera')
df2.pop('rendimientoUrbano')
```

Comienzo a categorizar las columnas en este estado, reviso el primer campo a categorizar


```python
df2.norma.unique()
```

```python
norma= df2.pop('norma')
#norma
df2['EURO V'] = (norma == 'EURO V')*1.0
df2['EURO VI'] = (norma == 'EURO VI')*1.0
df2['TIER 3 B50'] = (norma == 'TIER 3 B50')*1.0
df2['TIER 2 B5'] = (norma == 'TIER 2 B5')*1.0
df2['TIER 2 B4'] = (norma == 'TIER 2 B4')*1.0
df2['TIER 3 B70'] = (norma == 'TIER 3 B70')*1.0
df2['TIER 2 B6'] = (norma == 'TIER 2 B6')*1.0
df2['TIER 2 B8'] = (norma == 'TIER 2 B8')*1.0
df2['TIER 2 B7'] = (norma == 'TIER 2 B7')*1.0
df2['Tier 3 B125'] = (norma == 'Tier 3 B125')*1.0
df2['TIER 3 B125'] = (norma == 'TIER 3 B125')*1.0
df2['TIER 3 B30'] = (norma == 'TIER 3 B30')*1.0

df2.tail()
```

```python
df2.tipoVehiculo.unique()
```

```python
tipoVehiculo= df2.pop('tipoVehiculo')
#tipoVehiculo
df2['Hatch Back'] = (tipoVehiculo == 'Hatch Back')*1.0
df2['Sedán'] = (tipoVehiculo == 'Sedán')*1.0
df2['Station Wagon'] = (tipoVehiculo == 'Station Wagon')*1.0
df2['Cabriolet'] = (tipoVehiculo == 'Cabriolet')*1.0
df2['Coupé'] = (tipoVehiculo == 'Coupé')*1.0
df2['Convertible'] = (tipoVehiculo == 'Convertible')*1.0
df2['Furgón'] = (tipoVehiculo == 'Furgón')*1.0
df2['Jeep'] = (tipoVehiculo == 'Jeep')*1.0
df2['Limusina'] = (tipoVehiculo == 'Limusina')*1.0
df2['Roadster'] = (tipoVehiculo == 'Roadster')*1.0
df2['Camioneta'] = (tipoVehiculo == 'Camioneta')*1.0
df2['Minivan'] = (tipoVehiculo == 'Minivan')*1.0
df2['Furgón Cerrado'] = (tipoVehiculo == 'Furgón Cerrado')*1.0
df2['Minibus'] = (tipoVehiculo == 'Minibus')*1.0
df2['Microvan'] = (tipoVehiculo == 'Microvan')*1.0
df2.tail()
```

```python
df2.traccion.unique()
```

```python
traccion= df2.pop('traccion')
#traccion
df2['4x2'] = (traccion == '4x2')*1.0
df2['4x4'] = (traccion == '4x4')*1.0
df2.tail()
```

```python
df2.transmision.unique()
```

```python
transmision= df2.pop('transmision')
#traccion
df2['M'] = (transmision == 'M')*1.0
df2['A'] = (transmision == 'A')*1.0
df2.tail()
```
